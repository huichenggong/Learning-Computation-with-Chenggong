```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,2,100)
y = (x-1.5)**2
plt.figure(1,dpi=300)
plt.plot(x,y)
plt.yticks([])
plt.savefig("bond.png")
```

```python
x = np.linspace(89,129,100)
y = (x-109)**2
plt.figure(1,dpi=300)
plt.plot(x,y)
plt.yticks([])
plt.savefig("angle.png")
```

```python
x = np.linspace(0,360,500)
y = np.sin(np.deg2rad(x*3 + 90))
plt.figure(1,dpi=300)
plt.plot(x,y)
plt.yticks([])
plt.savefig("dihe.png")
```

```python
x = np.linspace(3,8,200) # A
#From Gromacs Amber ffnonbonded.itp
#K           19      39.10    0.0000  A   3.03796e-01  8.10369e-01  ; Joung2008 parameters for TIP3P
sigma = 3.03796e-01 *10
epsilon = 8.10369e-01 / 4.18 
y = 4* 8.10369e-01 * ((sigma/x)**12-(sigma/x)**6)
plt.figure(1,dpi=300)
plt.plot(x,y)
plt.xlim(-0.1,8)
plt.savefig("vdw.png")
```

```python
x = np.linspace(3,8,200) # A
#https://manual.gromacs.org/documentation/2019/reference-manual/definitions.html#defunits
#138.935458
y = 138.935458 * 1*1/((x/10)**2) /4.18
plt.figure(1,dpi=300)
plt.plot(x,y)
plt.xlim(-0.1,8)
plt.savefig("Coulomb.png")
```
