```python
import matplotlib.pyplot as plt
```

```python
name = ["mdtraj", "mdanalysis", "pytraj", "gmx distance", "C++"]
time = [29.341, 16.483, 16.258, 24.233, 15.635]
```

```python
plt.figure(1, dpi=600)
bars = plt.bar(name, time, color='blue', alpha=0.5)

plt.ylabel('Time (s)')
plt.title('Execution Time of 5 distance measurement')

for bar, t in zip(bars, time):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/4.0, yval, f"{t:.1f} s", va='bottom')  # va: vertical alignment
plt.savefig("01-exe_time.png")
```
